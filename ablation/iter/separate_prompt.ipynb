```python
path = "alpaca_llama70b_iteration_2-iter-filtered-separated.jsonl"
df = pd.read_json(path, lines=True)
```

```python
def parse_response(response):
    parsed = []
    for i in range(1, 10):
        # search for the instruction i
        instruction = response.find(f"Instruction {i}:")
        if instruction == -1:
            instruction = response.find(f"Split instruction {i}:")
        output = response.find(f"Response {i}:")
        if instruction == -1 or output == -1:
            break
        instruction = response[instruction+len(f"Instruction {i}:"):output].strip()
        next_instruction = response.find(f"Instruction {i+1}:")
        if next_instruction == -1:
            output = response[output+len(f"Response {i}:"):].strip()
        else:
            output = response[output+len(f"Response {i}:"):next_instruction].strip()
        parsed.append((instruction, output))
    if parsed:
        return parsed
    # search for Instruction N and Response N
    instruction_idx = response.find("Instruction N:")
    output_idx = response.find("Response N:")
    loop = 0
    while (instruction_idx != -1) and (output_idx != -1):
        # if loop > 10:
        #     break
        loop += 1
        instruction = response[instruction_idx+len("Instruction N:"):output_idx].strip()
        next_instruction = response.find("Instruction N:", output_idx)
        if next_instruction == -1:
            output = response[output_idx+len("Response N:"):].strip()
        else:
            output = response[output_idx+len("Response N:"):next_instruction].strip()
        parsed.append((instruction, output))
        response = response[output_idx+len("Response N:"):]
        instruction_idx = response.find("Instruction N:")
        output_idx = response.find("Response N:")
    
    return parsed
```

```python
from tqdm import tqdm
tqdm.pandas()
df['parsed'] = df['response'].progress_apply(parse_response)
```

```python
# explode the parsed column
df_explode = df.explode('parsed')
# extract the instruction and output by [0] and [1]
df_explode['instruction'] = df_explode['parsed'].apply(lambda x: x[0])
df_explode['output'] = df_explode['parsed'].apply(lambda x: x[1])
# drop the option, option_0, prompt, response and parsed columns
df_explode = df_explode.drop(columns=['option', 'option_0', 'prompt', 'response', 'parsed'])
df_explode
```

```python
not_filtered = pd.read_json("alpaca_llama70b_iteration_2-iter-not-filtered.jsonl", lines=True)
not_filtered = not_filtered.drop(columns=['option', 'option_0'])
# concatenate the filtered and not filtered
df_concat = pd.concat([df_explode, not_filtered])
df_concat
```

```python
# save as jsonl
df_concat.to_json("alpaca_llama70b_iteration_2-separated.jsonl", orient='records', lines=True)
```
